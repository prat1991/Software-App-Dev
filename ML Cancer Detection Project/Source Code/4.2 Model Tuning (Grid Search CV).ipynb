<h1 style="color:red"> 
4. Feature Selected Matrix
</h1>

```python
X_BE=dataset.iloc[:,[0,1,4,6,10,14]].values #stores all the rows of columns 0,1,4,6,10 & 14 into a new matrix called X_BE
```

<h1 style="color:red"> 
4.5 Splitting the dataset into training and testing datasets
</h1>

```python
#Splitting the dataset into 80% training dataset and 20% testing dataset from the cleansed dataset
# random state = 0 ensures that the data is not being split randomly everytiem
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_BE,y,test_size=0.2, random_state=0)

print('\nDataset sizes')
print('\t X_train size:', X_train.shape)
print('\t y_train size:', y_train.shape)
print('\t X_test size:', X_test.shape)
print('\t y_test size:', y_test.shape)
```

<h1 style="color:red"> 
5. Normalizing the training and testing datasets
</h1>

```python
#Before we create our classifier, we have to normalize the data (feature scaling) using the
#Scikit-Learn preprocessing package's StandardScalar function .
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)# fitting & transforming the training dataset
X_test = sc_x.transform(X_test) # transforming the testing dataset

print('\nDataset sizes')
print('\t X_train size:', X_train.shape)
print('\t X_test size:', X_test.shape)
```

<h1 style="color:red"> 
6. Model Hyperparameter Tuning (Grid Search CV)
</h1>

<h2 style="color:red"> 
6.1 SVM (Grid Search CV)
</h2>

```python
#importing functions
from sklearn.model_selection import GridSearchCV
from sklearn import svm

#creating the model object
model=svm.SVC( gamma='auto')

#defining the hyperparameters 
#(15 combinations of hyperparameters)
params = {'C': [1,2,3,4,5], 
          'kernel': ['linear','rbf']}

#Making the grid search model
model = GridSearchCV(model, param_grid=params, n_jobs=-1, cv=10)

#training the gridsearch model
model.fit(X_train,y_train)

#displaying the tuned hyperparameters of the trained gridsearch model
#(out of 150 different model combinations)
print("\nSVM Tuned Hyperparameters:\n",model.best_params_)

#testing the gridsearch model
y_prediction=model.predict(X_test)

#importing functions
from sklearn import metrics

#displaying the grid search models accuracy
acc1=metrics.accuracy_score(y_prediction,y_test)
print("\nSVM Accuracy(rate of making good predictions):\n",acc1)

#displaying the grid search models misclassification rate
MCR1=1-acc1
print("\nSVM Misclassification(rate of making bad predictions):\n ",MCR1)

#displaying the grid search models confusion matrix
cm = metrics.confusion_matrix(y_prediction,y_test)
xlabels=['No CHD','CHD']
ylabels=['No CHD','CHD']
df_cm = pd.DataFrame(cm,index=xlabels,columns=ylabels)
df_cm.index.name = 'Actual Class'
df_cm.columns.name = 'Predicted Class'
heatmap_cm=sns.heatmap(df_cm,annot=True,cmap='Blues',fmt='g')
heatmap_cm.set_title("SVM (Grid Search CV)")
plt.figure(figsize = (1,1))
sns.set(font_scale=1)
```

<h2 style="color:red"> 
6.2 Logistic Regression(Grid Search CV)
</h2>

```python
#importing functions
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

logregpipe = Pipeline([('scale', StandardScaler()),
                   ('logreg',LogisticRegression(multi_class="multinomial",solver="lbfgs"))])

# Gridsearch to determine the value of C  
#(5 combinations of hyperparameters)
param_grid = {'logreg__C':[1,25,50,75,100]}

#creating the model object  
#used pipelines here to show a different way of doing coding
model = GridSearchCV(logregpipe,param_grid,cv=10,return_train_score=True)
model.fit(X_train,y_train)
print("Logistic Regression Tuned Hyperparameters:",model.best_params_)

#training the gridsearch model
bestlogreg = model.best_estimator_
bestlogreg.fit(X_train,y_train)

#displaying the tuned hyperparameters of the trained gridsearch model
#(out of 50 different model combinations)
bestlogreg.coef_ =bestlogreg.named_steps['logreg'].coef_

#displaying the grid search models accuracy 
acc2=bestlogreg.score(X_train,y_train)
print("\nLogistic Regression Accuracy(rate of making good predictions):\n",acc2)

#displaying the grid search models misclassification rate
MCR2=1-acc2
print("\nLogistic Regression Misclassification(rate of making bad predictions)\n ",MCR2)

#testing the gridsearch model
y_prediction=model.predict(X_test)

#importing functions
from sklearn import metrics

#displaying the grid search models confusion matrix
cm =metrics.confusion_matrix(y_prediction,y_test)
xlabels=['No CHD','CHD']
ylabels=['No CHD','CHD']
df_cm = pd.DataFrame(cm,index=xlabels,columns=ylabels)
df_cm.index.name = 'Actual Class'
df_cm.columns.name = 'Predicted Class'
heatmap_cm=sns.heatmap(df_cm,annot=True,cmap='Blues',fmt='g')
heatmap_cm.set_title("Logistic Regression (Grid Search CV)")
plt.figure(figsize = (1,1))
sns.set(font_scale=1)
```

<h2 style="color:red"> 
6.4 Random Forest (Grid Search CV)
</h2>

```python
#importing functions
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

#creating the model object
model=RandomForestClassifier()

#defining the hyperparameters
# (10 combinations of hyperparameters)
params = {'criterion':['gini','entropy'],
          'n_estimators':[1,25,50,75,100] # number of trees
         }

#Making the grid search model
model = GridSearchCV(model, param_grid=params, n_jobs=-1, cv=10)

#training the gridsearch model
model.fit(X_train,y_train)

#displaying the tuned hyperparameters of the trained gridsearch model
#(out of 100 different model combinations)
print("\nRandom Forest Tuned Hyperparameters:\n",model.best_params_)

#testing the gridsearch model
y_prediction=model.predict(X_test)

#importing functions
from sklearn import metrics

#displaying the grid search models accuracy
acc4=metrics.accuracy_score(y_prediction,y_test)
print("\nRandom Forest Accuracy(rate of making good predictions):\n",acc4)

#displaying the grid search models misclassification rate
MCR4=1-acc4
print("\nRandom Forest Misclassification(rate of making bad predictions):\n ",MCR4)

#displaying the grid search models confusion matrix
cm = metrics.confusion_matrix(y_prediction,y_test)
xlabels=['No CHD','CHD']
ylabels=['No CHD','CHD']
df_cm = pd.DataFrame(cm,index=xlabels,columns=ylabels)
df_cm.index.name = 'Actual Class'
df_cm.columns.name = 'Predicted Class'
heatmap_cm=sns.heatmap(df_cm,annot=True,cmap='Blues',fmt='g')
heatmap_cm.set_title("Random Forest (Grid Search CV)")
plt.figure(figsize = (1,1))
sns.set(font_scale=1)
```

```python
#importing functions
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

#creating the model object
model = KNeighborsClassifier()

#defining the hyperparameters
# (25 combinations of hyperparameters)
params = {'n_neighbors':[1,25,50,75,100], # number of neigobors
          'leaf_size':[1,25,50,75,100]
         }

#Making the grid search model
model = GridSearchCV(model, param_grid=params, n_jobs=-1, cv=10)

#training the gridsearch model
model.fit(X_train,y_train)

#displaying the tuned hyperparameters of the trained gridsearch model
# (out of 250 different model combinations)
print("\nKNN Tuned Hyperparameters:\n",model.best_params_)

#testing the gridsearch model
y_prediction=model.predict(X_test)

#importing functions
from sklearn import metrics

#displaying the grid search models accuracy
acc5=metrics.accuracy_score(y_prediction,y_test)
print("\nKNN Accuracy(rate of making good predictions):\n",acc5)

#displaying the grid search models misclassification rate
MCR5=1-acc5
print("\nKNN Misclassification(rate of making bad predictions):\n ",MCR5)

#displaying the grid search models confusion matrix
cm = metrics.confusion_matrix(y_prediction,y_test)
xlabels=['No CHD','CHD']
ylabels=['No CHD','CHD']
df_cm = pd.DataFrame(cm,index=xlabels,columns=ylabels)
df_cm.index.name = 'Actual Class'
df_cm.columns.name = 'Predicted Class'
heatmap_cm=sns.heatmap(df_cm,annot=True,cmap='Blues',fmt='g')
heatmap_cm.set_title("KNN (Grid Search CV)")
plt.figure(figsize = (1,1))
sns.set(font_scale=1)
```
