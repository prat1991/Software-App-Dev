<h1 style="color:red"> 
6. Model Hyperparameter Tuning (KFolds Cross Val)
</h1>

```python
#-------------------------------------KFolds Cross Val Function-------------------------
from sklearn.model_selection import KFold

def run_kfold(obj):
    kf = KFold(n_splits=10)
    outcomes = []
    fold = 0
    for train_index, test_index in kf.split(X):
        fold += 1
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        obj.fit(X_train, y_train)
        accuracy = obj.score(X_test, y_test)
        outcomes.append(accuracy)
        print("Fold {0} accuracy: {1}".format(fold, accuracy))     
    mean_outcome= np.mean(outcomes)
    print("Mean Accuracy: {0}".format(mean_outcome)) #KFold cross validation
    return(mean_outcome)
```

<h2 style="color:red"> 
6.1 Linear SVM (KFolds Cross Val)
</h2>

```python
from sklearn import svm # Importing model library
obj1 = svm.SVC(kernel='linear', gamma="auto") # creating a model object
acc1=run_kfold(obj1) #KFolds Cross Val Function
```
